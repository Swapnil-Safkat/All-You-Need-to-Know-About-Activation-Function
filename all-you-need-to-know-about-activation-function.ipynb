```python
import matplotlib.pyplot as plt
import numpy as np
def plot(function,x,name):
    y = [function(i) for i in x]

    fig, axs = plt.subplots(1, 2)
    fig.suptitle(name+" activation function")
    fig.set_figwidth(12)
    fig.set_figheight(4)

    # Plot data in the first subplot
    axs[0].plot(x, x)
    axs[0].set_title('Actual')
    axs[0].grid()
    axs[0].legend()
    # Plot data in the second subplot
    axs[1].plot(x, y)
    axs[1].set_title('After Activation function')
    axs[1].grid()
    axs[1].legend()
    plt.show()
```

```python
def linear(x):
    return x
x = np.linspace(-10, 10, 400)
plot(linear,x,'linear')
```

```python
def step(x):
    return 10 if x >= 0 else -10
x = np.linspace(-10, 10, 400)
plot(step,x,'step')
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
x = np.linspace(-10, 10, 400)
plot(sigmoid,x,'sigmoid')
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
x = np.linspace(-10, 10, 400)
plot(tanh,x,'tanh')
```

```python
def relu(x):
    return max(0,x)
x = np.linspace(-10, 10, 400)
plot(relu,x,'relu')
```

```python
def leaky(x):
    return np.where(x > 0, x, 0.05 * x)
x = np.linspace(-10, 10, 400)
plot(leaky,x,'leaky relu')
```

```python
def elu(x):
    return np.where(x > 0, x, 0.8*(np.exp(x)-1))
x = np.linspace(-10, 10, 400)
plot(elu,x,'elu')
```
